```python
from PIL import Image
import requests
from io import BytesIO
import pytesseract
from IPython.display import display
import tempfile
import imagehash
from collections import Counter
import json
from multiprocessing import Pool, Queue
import os

# 二值化處理 (處理後反而正確率下降.. 所以不用)
def preprocess(im):
    pixdata = im.load()
    for y in range(im.size[1]):
        for x in range(im.size[0]):
            #if pixdata[x,y][0]>150 and pixdata[x,y][1]>150 and pixdata[x,y][2]>150:
            #    pixdata[x,y]=(255, 255, 255, 0)
            #else:
            pixdata[x,y]=( pixdata[x, y][0], 0, 0, 0)
    return im


# 取得圖片 by url
def get_image(img_url):
    response = requests.get(img_url)
    # print(response.url)
    img = Image.open(BytesIO(response.content))
    return img

# 預測單行
def detect(img):
    il = img.crop((334, 561, 372, 585))
    ir = img.crop((335, 603, 371, 630))
    #il, ir = preprocess(il), preprocess(ir)
    #display(il)
    #display(ir)
    img_hashs = (imagehash.average_hash(il, hash_size=10), imagehash.average_hash(ir, hash_size=10))
    points = pytesseract.image_to_string(il, config='--psm 7'),pytesseract.image_to_string(ir, config='--psm 7')

    return [points, img_hashs]

# 確認 logo 
def is_activity(img):
    t = img.crop((149, 569, 195, 636))
    ori = imagehash.hex_to_hash('7cfef7586c003c00')
    new_hash = imagehash.average_hash(t)
    return ori - new_hash <= 3

```

```python
# 影片切片
def video_split_by_point(v, folder='result'):
    data = open('{}/{}'.format(folder, v.id)).read()
    data = json.loads(data)
    if not data:
        return False
    tmp_t = data[0][1]
    tmp_p = data[0][0]
    counter = Counter()
    infos = []
    for d in data:
        l, r = d[0]
        l = re.sub("[,.;‘)}:]", "", l).replace('A', '4').strip()
        r = re.sub("[,.;‘)}:]", "", r).replace('A', '4').strip()

        t = d[1]
        h = d[2]
        diff = t - tmp_t
        tmp_t = t
        counter.update([tuple([l, r])])
        infos.append([(l,r), t, diff, d[2], tuple(tmp_p)])
        tmp_p = (l, r)

    tmp_p = None
    old = []
    h_c = 0
    for i in infos:
        try:
            l, r = i[0]
            l, r = int(l), int(r)
            # 一場比賽應該不會超過 300 分..
            assert l<300 and r < 300
            # 如果該分數出現次數不夠, 判定誤判
            if (l, r) not in old and tmp_p != (l, r) and counter[i[0]] > 4:
                tmp_p = (l, r)
                old.append(tmp_p)
                if i[2] < 30:
                    v.slice("{}:{}".format(l, r), i[1] - 9, i[1], meta=json.dumps(i))
            
            # 如果中場時間小於60秒 判定是hightlight
            if i[2] > 10 and i[2] < 60:
                h_c += 1
                v.slice("{}:{}:hight_light{}".format(l, r, h_c), i[1] - i[2] + 3, i[1] - 2, meta=json.dumps(i))
                print(tmp_p, i)
        except:
            pass
    tmp = (0, 0)
    
    # 合併事件
    events = get_events(live_map[v.name + ".mp4"])
    for event in events:
        print(event)
        l, r = int(event['homeScore']), int(event['awayScore'])
        if (l, r) != tmp:
            tmp = (l, r)
            s = VideoScene.objects.filter(video=v, text="{}:{}".format(l, r))
            if s:
                s = s[0]
            else:
                continue
            s.text = "[{}:{}]{}".format( l, r, event.get('msg', {}).get('content', ''))
            # 罰球
            if "罚" in s.text:
                s.start = s.end - 30
            meta = json.loads(s.meta)
            meta.append(event)
            s.meta = json.dumps(meta)
            s.save()
    return True
```
