```python
import requests
import re
from bs4 import BeautifulSoup

live_url_pattern = re.compile('https://www.jleague.jp/match/\w+/\d+/\d+/live/(#.*)?')
time_pattern = re.compile('([前後]半)(\d+)分')

# 驗證連結
def check_and_get_base_url(live_url):
    match = live_url_pattern.match(live_url)
    if match:
        return live_url.split('/live/')[0]
    else:
        raise Exception('url not support')

# 取得比賽事件 api url
def live_text_api(live_url):
    base_url = check_and_get_base_url(live_url)
    api = '/ajax_livetxt'
    api_url = base_url + api
    return api_url

# 取得比賽結果
def get_result(live_url):
    check_and_get_base_url(live_url)
    resp = requests.get(live_url)
    soup = BeautifulSoup(resp.text)
    score_elem = soup.select('.leagScoreColumn')[0]
    left, right = score_elem.select('.leagAccTeam p')
    left_team = left.text
    right_team = right.text
    
    left, right = score_elem.select('.leagAccTeam span')
    left_full = left.text
    right_full = right.text

    
    left_score = int(score_elem.select('.leagLeftScore')[0].text)
    right_score = int(score_elem.select('.leagRightScore')[0].text)
    title_context = soup.select('.matchBoxTitleArea')[0].text
    
    section = re.search('第(\d+)節', title_context).groups()[0]
    time = re.search(".*年.*", title_context).group().strip()
    
    return {
        "left": left_team,
        "left_full": left_full,
        "right": right_team,
        "right_full": right_full,
        "left_score":left_score,
        "right_score":right_score,
        "section": section,
        "type": "J1",
        "time": time
    }
    

# 取得比賽事件
def get_event_points(live_url):
    api = live_text_api(live_url)
    resp = requests.get(api)
    soup = BeautifulSoup(resp.text)
    event_elems = reversed(soup.select('.matchSpotListArea li'))

    data = {}
    events = {"first": [], "second": []}
    second_start = None
    idx = 0
    for elem in event_elems:
        event = {}
        game_time = elem.select('.spotLeftTxt')[0].text
        session, time = time_pattern.match(game_time).groups()

        time = int(time) * 60
        event['time'] = time
        event['text'] = elem.select('.spotRightTxt')[0].text
        event['team'] = elem.select('.spotCenterTeam')[0].text
        event['score'] = bool(elem.select('.goalVideoThum'))
        event['idx'] = idx
        idx += 1

        if event['text'].startswith('前半終了。'):
            second_start = time

        if session.startswith('前半'):
            events['first'].append(event)
        if session.startswith('後半'):
            events['second'].append(event)

    data['events'] = events
    data['second_start'] = second_start
    return data
```

```python
from material import hls
from material.models import Video, VideoScene
import json

def dumps(data):
    return json.dumps(data, ensure_ascii=False, indent=2)

j1_indexs = [l.split(',') for l in j1_index_data.split('\n')]
for name, live_url, first, second in j1_indexs:
    url = "https://storage.googleapis.com/livingbio-library/j1/2017hls/{}/video.m3u8".format(name)
    print(url)
    
    # 上半場開始時間
    f_m, f_s = first.split('.')
    first_start = int(f_m) * 60 + int(f_s)

    # 下半場開始時間
    s_m, s_s = second.split('.')
    second_start = int(s_m) * 60 + int(s_s)

    m3u8 = hls.M3U8.from_url(url)
    delta = 60

    # 取得整場訊息
    result = get_result(live_url)

    # 取得事件時間資訊
    scenes_info = get_event_points(live_url)

    # 影片
    source_video = Video.sync(name, url)
    source_video.meta = dumps(result)
    source_video.save()

    # 上半場 scenes
    idx = 0
    for event in scenes_info['events']['first']:
        event['idx'] = idx
        t = event['time']
        st = max(0, t-delta) + first_start
        ed = t + delta + first_start
        new = m3u8.slince(st, ed)
        VideoScene.objects.create(video=source_video, start=st, end=ed, text=event['text'], meta=dumps(event))

    # 下半場 scenes
    idx = 0
    for event in scenes_info['events']['second']:
        t = event['time']
        event['idx'] = idx
        st = max(0, t-delta) + second_start
        ed = t + delta + second_start
        new = m3u8.slince(st, ed)
        VideoScene.objects.create(video=source_video, start=st, end=ed, text=event['text'], meta=dumps(event))
```
