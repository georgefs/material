```python
# 增加 隊伍 tags
def add_team_tags(s):
    try:
        info = json.loads(s.meta)
        p0 = [i for i in info[4]]
        p1 = [i for i in info[0]]
        print(p0, p1)
        if p0[0] == p1[0]:
            if p0[1] != p1[1]:
                print('away', p0, p1)
                tag, _ = Tag.objects.get_or_create(name="{}_score".format(info[5]['awayTeam']))
                s.tags.add(tag)
        elif p0[1] == p1[1]:
            if p0[0] != p1[0]:
                print('home', p0, p1)
                tag, _ = Tag.objects.get_or_create(name="{}_score".format(info[5]['homeTeam']))
                s.tags.add(tag)
    except Exception as e:
        #print(info[4], info[0])
        print(e)
        pass
```

```python
# event 分類
event_mappings = \
{
"灌篮":["暴扣","扣"],
"三分":["三分","3分"],
"上篮":["上篮"],
"快攻":["快攻"],
"中距离":["中距离","中投","踩线"],
"篮下":["篮下强攻","篮下","放进","强打"],
"射篮":["跳投","抛投","投了","投一发","后仰","抛射","骑射","骑马射箭","强投","射一发"],
"其他事件":["拉杆","擦板","罚","补篮","挑篮","打板","勾手","反篮","干拔","反击","空切","跑投","强起","放篮","半截篮","放球","舔篮","放筐","上空篮","补进","打进","补中","两分"],
}

def add_event_tags(s):
    video_tags = []
    for key, values in event_mappings.items():
        tag, _ = Tag.objects.get_or_create(name="{}_event".format(key))
        for value in values:
            # 有事件存在就不需要標其他事件
            if video_tags and tag.name == '其他事件':
                break

            if value in s.text:
                video_tags.append(tag)
                break
        
    s.tags.add(*video_tags)
    return True
```

```python
from material.models import VideoScene
# 暫時先判斷是 [ 開頭的才為得分事件
video_scenes = VideoScene.objects.filter(text__startswith='[')

for scene in video_scenes:
    add_player_tags(scene)
    add_team_tags(scene)
    add_event_tags(scene)
    print(scene.id, scene, scene.tags.all())
```
