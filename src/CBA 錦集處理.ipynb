```python
# 球員錦集
from material.models import VideoScene, Video, Collection
from django.db.models import Count
import random
import json
c = 0
data = []
for video in Video.objects.filter(id__gte=82):
    video_name = video.name.replace('[全场]_', '')
    vs = VideoScene.objects.filter(video=video). \
        filter(tags__name__contains='score_player').exclude(text__contains='罚').values('tags__name'). \
        annotate(Count('tags__name')).filter(tags__name__count__gte=4)
    for info in vs:
        #print(info)
        tag_name = info['tags__name']
        count = info['tags__name__count']
        scenes = VideoScene.objects.filter(video=video, tags__name=tag_name).exclude(text__contains='罚').distinct()
        ids = []
        tmp = [i.pk for i in scenes]
        permary_events = list(scenes.filter(tags__name__contains='灌篮_event').distinct())
        other_events = list(scenes.exclude(tags__name__contains='灌篮_event').distinct())
        loss = 8 - len(permary_events) 
        if loss < len(other_events):
            fill_events = random.sample(other_events, k=loss)
        else:
            fill_events = other_events
        events = permary_events + fill_events
        ids = sorted([e.id for e in events])
        score_team = events[0].tags.filter(name__endswith='_score')[0].name.replace('_score', '')
        c+=1
        # print(VideoScene.objects.filter(id__in=ids).filter(text__contains='罚'))
        data.append([video_name, score_team, tag_name, ids])
        print(json.dumps([video_name, score_team, tag_name, ids], ensure_ascii=False))
        player_name = tag_name.replace('_score_player', '')
        cv, _ = Collection.objects.get_or_create(name="【{}】全場精彩表現 {}".format(player_name, video_name))
        cv.scenes = ",".join([str(i) for i in ids])
        cv.meta = json.dumps([video_name, score_team, tag_name, ids])
        cv.save()                                        
        
print(c)
```

```python
# 隊伍錦集

from django.db.models import Count
c = 0 
for video in Video.objects.filter(id__gte=82):
    video_name = video.name.replace('[全场]_', '')
    vs = VideoScene.objects.filter(video=video). \
        filter(tags__name__endswith='_score').exclude(tags__name='其他事件_event').values('tags__name'). \
        annotate(Count('tags__name')).order_by('-tags__name__count')
    for info in vs:
        #print(info)
        tag_name = info['tags__name']
        count = info['tags__name__count']
        scenes = VideoScene.objects.filter(video=video, tags__name=tag_name).exclude(text__contains='罚').distinct()
        ids = []
        permary_events = list(scenes.filter(tags__name__contains='灌篮_event').distinct())
        other_events = list(scenes.exclude(tags__name__contains='灌篮_event').distinct())
        loss = 10 - len(permary_events) 
        if loss < len(other_events):
            fill_events = random.sample(other_events, k=loss)
        else:
            fill_events = other_events
        events = permary_events + fill_events
        ids = sorted([e.id for e in events])
        c+=1
        print(json.dumps([video_name, tag_name, ids], ensure_ascii=False))
        # print(VideoScene.objects.filter(id__in=ids).filter(text__contains='罚'))
        team_name = tag_name.replace('_score', '')
        cv, _ = Collection.objects.get_or_create(name="【{}】全场精华 {}".format(team_name, video_name))
        cv.scenes = ",".join([str(i) for i in ids])
        cv.meta = json.dumps([video_name, tag_name, ids])
        cv.save()                                        

print(c)
```

```python
# 事件錦集

from django.db.models import Count
c = 0
mapping = {
    u"三分_event": u"【三分球】本场三分球精华",
    u"灌篮_event": u"【灌篮】本场火热灌篮精华",
    u"快攻_event": u"【快攻】盘点本场精彩快攻",
    u"篮下_event": u"【禁区强打】本场禁区对抗精彩片段",
    u"上篮_event": u"【禁区强打】本场禁区对抗精彩片段",
    u"射篮_event": u"【射篮】本场射篮精华",
    u"中距离_event": u"【射篮】本场射篮精华",
}

for video in Video.objects.filter(id__gte=82):
    video_name = video.name.replace('[全场]_', '')
    vs = VideoScene.objects.filter(video=video). \
        filter(tags__name__contains='_event').exclude(tags__name='其他事件_event').values('tags__name'). \
        annotate(Count('tags__name')).order_by('-tags__name__count')
    
    for info in vs:
        #print(info)
        tag_name = info['tags__name']
        count = info['tags__name__count']
        scenes = VideoScene.objects.filter(video=video, tags__name=tag_name).exclude(text__contains='罚').distinct()
        ids = []
        scenes = list(scenes)
        
        if 8 < len(scenes):
            scenes = random.sample(scenes, k=8)
        else:
            scenes = scenes
        events = scenes
        ids = sorted([e.id for e in events])
        if len(ids) == 1 or (len(ids) < 5 and tag_name != '灌篮_event'):
            continue
        
        c+=1
        print(json.dumps([video_name, tag_name, ids], ensure_ascii=False))
        
        # print(VideoScene.objects.filter(id__in=ids).filter(text__contains='罚'))
        team_name = tag_name.replace('_score', '')
        cv, _ = Collection.objects.get_or_create(name="{} {}".format(mapping[tag_name], video_name))
        cv.scenes = ",".join([str(i) for i in ids])
        cv.meta = json.dumps([video_name, tag_name, ids])
        cv.save()                                        
        # print(VideoScene.objects.filter(id__in=ids).filter(text__contains='罚'))
```
